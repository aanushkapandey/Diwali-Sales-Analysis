```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True, errors='ignore')
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Relationship_Status'}, inplace=True)
```

```python
df.head()
```

# **Exploratory Data Analysis**

**Gender**

```python
df.columns
```

```python
ax= sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gender)

plt.show()
```

for above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men.

**AGE**

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data= sales_age)

plt.show()
```

from above graph we can see that the most buyers are of the age group  between 26-35 years female

**STATE**

```python
df.columns
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Orders')

plt.show()
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set({'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Amount')

plt.show()
```

from above graph we can see that Uttar Pradesh, Maharashtra, Karnataka. 

**Marital Status**

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Relationship_Status')
sns.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_relation= df.groupby(['Relationship_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_relation, x='Relationship_Status', y='Amount', hue='Gender')

plt.show()
```

fro above graph we can see that most of the buyers are married women and they have high purchasing power too.

**OCCUPATION**

```python
df.columns
```

```python
ax=sns.countplot(data=df, x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_occupation= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_occupation, x='Occupation', y='Amount')

plt.show()
```

from above graph we can see that most buyers are from IT, Aviation and Healthcare sector.


**PRODUCT**

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Product_Category')
sns.set(rc={'figure.figsize':(20, 5)})

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_product=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_product, x= 'Product_Category', y='Amount')

plt.show()
```

from above graph we can see that most of the sold products are food, cloths and electro

```python
sales_product= df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_product, x='Product_ID', y='Amount')

plt.show()
```
